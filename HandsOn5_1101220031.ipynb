```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
fr = pd.read_csv("FuelConsumptionCo2.csv")
fr.head()
```

```python
fr.describe()
```

```python
cry = fr [['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
cry.head(10)
```

```python
cry[['CYLINDERS', 'ENGINESIZE', 'CO2EMISSIONS', 'FUELCONSUMPTION_COMB']].plot.hist(
    subplots=True, layout=(2, 2), figsize=(10, 8), bins=20, alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.scatter(cry['FUELCONSUMPTION_COMB'], cry['CO2EMISSIONS'], color='navy')
plt.xlabel("FUELCONSUMPTION_COMB")
plt.ylabel("Emission")
plt.show()
```

```python
cry.plot.scatter(x='ENGINESIZE', y='CO2EMISSIONS', color='maroon')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

```python
msk = np.random.rand(len(fr)) < 0.8
train = cry[msk]
test = cry[~msk]
```

```python
train.plot.scatter(x='ENGINESIZE', y='CO2EMISSIONS', color='green')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

```python
lr = LinearRegression()
X_train = np.asanyarray(train[['ENGINESIZE']])
y_train = np.asanyarray(train[['CO2EMISSIONS']])

lr.fit(X_train, y_train)

print('Coefficients:', lr.coef_)
print('Intercept:', lr.intercept_)
```

```python
train_x = np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])

regr = LinearRegression()
regr.fit(train_x, train_y)
```

```python
import matplotlib.pyplot as plt

plt.scatter(train["ENGINESIZE"], train["CO2EMISSIONS"], color='pink', label="Data asli")
plt.plot(train["ENGINESIZE"], regr.predict(train[["ENGINESIZE"]]), '-r', label="Regresi linier")
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.legend()
plt.grid(True)
plt.show()
```
